```python
import pandas as pd
from pandas import to_datetime as dt
import matplotlib.pyplot as plt
data={"Date": dt(["2004-01-03", "2004-02-08", "2004-03-10", "2004-04-25", "2004-05-26", "2004-06-24"]),
      "Value": [10,5,11,9,17,19]}
df = pd.DataFrame(data)
df.set_index("Date", inplace=True)
plt.figure(figsize=(10, 6))
plt.plot(df.index, df["Value"], marker='o')
plt.xlabel("Data")
plt.ylabel("Value")
plt.title("Time Series with seasonality")
plt.grid(True)
plt.show()
```
